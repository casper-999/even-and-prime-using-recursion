```python
import numpy as np
```

```python
random_int= np.random.randint(low=0,high=2000,size=1000)
all_numbers=list(random_int)  #function like pop doesnt exist in numpy arrays, so converting ndarray into list
```

```python
def if_prime(num,div):
    '''
    this function if the number is prime or not using recursion
   
    '''
    if num==2 or num==1:
        return True
    else:
        if div==num:
            return True
        if num%div==0:
            return False
        bool_value = if_prime(num,div+1)
    return bool_value
```

```python
#creating empty list
even=[]
prime=[]
```

```python

def even_prime_list(all_numbers,even,prime):
    '''
    this function will separate into even and prime into resp. lists 
     Args:
    even:list
    prime:list
    '''
    if len(all_numbers)==0:
        return prime,even
    if all_numbers[0]%2==0:
        even.append(all_numbers[0])
    else:
        if (if_prime(all_numbers[0],2)==True):
            prime.append(all_numbers[0])
    all_numbers.pop(0)
    prime_list,even_list=even_prime_list(all_numbers,even,prime)
    return prime_list,even_list        
```

```python
prime_list,even_list=even_prime_list(all_numbers,even,prime)
```
